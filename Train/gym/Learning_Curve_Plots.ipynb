```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_search = pd.read_csv('Final/RandomSearch_test.csv')
random_search.head()
```

```python
lc=random_search.plot.line(x="Number of Evaluations", y="Best Overall Speed", label="Random Search Best Overall Speed")
_=random_search.plot.line(x="Number of Evaluations", y="Speed", label="Iteration speed", ax=lc)
lc.set_title('Learning Curves')
lc.set_ylabel("Speed")
lc.set_ylim(0, 50)
lc.set_xlim(0,100)
plt.savefig("RandomSearchLearningCurve.png")
```

```python
hill_climber = pd.read_csv('Final/HillClimber_test.csv')
hill_climber.head()
```
